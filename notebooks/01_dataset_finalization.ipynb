```python
import pandas as pd

df = pd.read_csv("../data/all_india_cpi.csv")
df.head()
```

```python
df.info()
```

```python
df = df[df['Sector'] == 'Rural+Urban']
```

```python
df['Sector'].unique()
```

```python
df['Date'] = pd.to_datetime(
    df['Year'].astype(str) + '-' + df['Month'],
    format='%Y-%B'
)

df.set_index('Date', inplace=True)
```

```python
df['Date'] = pd.to_datetime(
    df['Year'].astype(str) + '-' + df['Month'],
    errors='coerce'
)
```

```python
df.set_index('Date', inplace=True)
```

```python
df.isna().sum()
```

```python
df.index = df.index.fillna(method='ffill')
```

```python
df = df[~df.index.isna()]
```
